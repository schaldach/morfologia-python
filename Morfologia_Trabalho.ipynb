```python
# importando as bibliotecas necessárias
#from https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_morphological_ops/py_morphological_ops.html
import cv2
import numpy as np
#caso for usar o Google Colab com a OpenCV, usar a lib abaixo
from google.colab.patches import cv2_imshow
```

```python
# inicializando as imagens, que devem ser carregadas dentro dos arquivos do colab, salvando a altura e largura das imagens
# e declarando o kernel que manipulará as imagens
img = cv2.imread('image.png',0)
img_opening = cv2.imread('image (1).png',0)
img_closing = cv2.imread('image (2).png',0)
altura, largura = img.shape
kernel = np.ones((5,5),np.uint8)
print(kernel)
print(altura, largura)
```

```python
# usando a função cv2.erode para aplicar a "erosão" na imagem e cv2.dilate para aplicar a "dilatação" na imagem
erosion = cv2.erode(img,kernel,iterations = 4)
dilation = cv2.dilate(img,kernel,iterations = 4)
```

```python
# usando a função cv2.morphologyEx para aplicar a diferença entre erosão e dilatação, erosão seguida de dilatação
# e a dilatação seguida de erosão
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
opening = cv2.morphologyEx(img_opening, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(img_closing, cv2.MORPH_CLOSE, kernel)
```

```python
# código que usa a função cv2_imshow para mostrar uma imagem como output do Python, no caso são mostradas
# todas as imagens usadas até agora
'''
#Caso usa com Python no próprio computador
cv2.imshow('in', img)
cv2.imshow('erosion_out', erosion)
cv2.imshow('dilation_out', dilation)
cv2.imshow('opening_out', opening)
cv2.imshow('closing_out', closing)
cv2.imshow('gradient_out', gradient)
'''
#Caso use o Google Colab
cv2_imshow(img)
cv2_imshow(erosion)
cv2_imshow(dilation)
cv2_imshow(opening)
cv2_imshow(closing)
cv2_imshow(gradient)
```
